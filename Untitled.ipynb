```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv('Dataset/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def genresconvert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(genresconvert)
```

```python
movies.head(2)
```

```python
movies['keywords'] = movies['keywords'].apply(genresconvert)
```

```python
def castconvert(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 4:
            l.append(i['name'])
        counter = counter +1
    return l
```

```python
movies['cast'] = movies['cast'].apply(castconvert)
```

```python
movies.head(2)
```

```python
def director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
    return l
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
def collapse(obj):
    l = []
    for i in obj:
        l.append(i.replace(" ", ""))
    return l
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
```

```python
movies.head(5)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head(1)
```

```python
new = movies.drop(columns=['overview', 'genres', 'keywords', 'keyword', 'cast', 'crew'])
```

```python
new = movies.drop(columns=['overview', 'genres', 'keywords', 'cast', 'crew'])
```

```python
new.head(1)
```

```python
new.iloc[0].tags
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
```

```python
new.iloc[0].tags
```

```python
new['tags'] = new['tags'].apply(lambda x:x.lower())
```

```python
new.iloc[0].tags
```

```python
new.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
new.head(1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
len(similarity[0])
```

```python
def recommend(movie):
    index = new[new['title']==movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key = lambda x:x[1])
    
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
